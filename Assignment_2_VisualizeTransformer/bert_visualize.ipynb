```python
from transformers import BertTokenizer, BertModel
from bertviz import head_view
import torch
```

```python
model = BertModel.from_pretrained('bert-base-uncased', output_attentions=True)
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
```

```python
sentence = "Transformers are amazing!"

# Tokenize with return_tensors and return token_type_ids
inputs = tokenizer(sentence, return_tensors="pt", return_token_type_ids=True)

# Save input tokens for visualization
input_ids = inputs['input_ids']
token_type_ids = inputs['token_type_ids']

tokens = tokenizer.convert_ids_to_tokens(input_ids[0])
```

```python
# Run the model and get attentions
with torch.no_grad():
    outputs = model(**inputs)

attentions = outputs.attentions  # Tuple of (num_layers,) with shape [1, num_heads, seq_len, seq_len]
```
